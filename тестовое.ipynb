```python
# Импортируем необходимые библиотеки
from Bio import SeqIO

# Загрузка файла из Google Drive
from google.colab import drive
drive.mount('/content/gdrive')
fasta_file = '/content/gdrive/MyDrive/GRCh38.d1.vd1.fa'

# Открытие FASTA-файла
with open(fasta_file, "r") as handle:
    # Чтение последовательностей из файла
    sequences = list(SeqIO.parse(handle, "fasta"))

# Вывод информации о последовательностях
for seq in sequences:
    print(f"Название последовательности: {seq.id}")
    print(f"Длина последовательности: {len(seq)}")
    print(f"Последовательность: {seq.seq}")
    print()
```

```python
import os
import pysam
from google.colab import drive

# Подключаемся к Google Drive
drive.mount('/content/drive')

# Путь к референсному геному
ref_genome_path = '/content/drive/MyDrive/GRCh38.d1.vd1.fa'

# Проверяем, что референсный геном доступен
if os.path.exists(ref_genome_path):
    print("Референсный геном найден.")
else:
    print("Референсный геном не найден. Проверьте путь.")
    exit()

# Создаем объект pysam.Fastafile для доступа к референсному геному
ref_genome = pysam.Fastafile(ref_genome_path)

# Путь к вашему FASTA-файлу
fasta_file_path = '/content/drive/MyDrive/FP_SNPs_10k_GB38_twoAllelsFormat.tsv'

# Создаем новый FASTA-файл для сохранения результатов
output_file_path = '/content/drive/MyDrive/FP_SNPs_10k_GB38_twoAllelsFormatWithRef.tsv'

# Функция для получения референсного аллеля
def get_ref_allele(chrom, pos):
    try:
        ref_allele = ref_genome.fetch(f'chr{chrom}', pos - 1, pos).upper()
        return ref_allele
    except KeyError:
        # Если позиция не найдена в референсном геноме, возвращаем пустую строку
        return ''

# Открываем исходный FASTA-файл и обрабатываем последовательности
with open(fasta_file_path, 'r') as fasta_file, open(output_file_path, 'w') as output_file:
    for line in fasta_file:
        if line.startswith('>'):
            # Получаем идентификатор последовательности
            seq_id = line.strip('>\n')
            chrom, pos = seq_id.split('_')
            pos = int(pos)

            # Получаем референсный аллель
            ref_allele = get_ref_allele(chrom, pos)

            # Записываем результат в выходной файл
            output_file.write(f"{seq_id}\tchr{chrom}\t{pos}\t{ref_allele}\t{line.strip('>')}\n")
        else:
            # Пропускаем последовательность, поскольку она уже учтена в предыдущей строке
            pass

# Закрываем объект pysam.Fastafile
ref_genome.close()

print("Скрипт завершен. Результаты сохранены в файле:", output_file_path)
```
